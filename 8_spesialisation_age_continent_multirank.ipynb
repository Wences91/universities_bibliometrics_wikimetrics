```python
import pandas as pd
import numpy as np
```

Function to remove special characters.

```python
def simplify(text):
    import unicodedata
    try:
        text = unicode(text, 'utf-8')
    except NameError:
        pass
    text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode("utf-8")
    return str(text)
```

Gini index function.

```python
def gini(x):
    total = 0
    x = np.array(x)
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))
```

# Leiden Ranking

```python
df_l = pd.read_csv('data/full_leiden_wikipedia_indicators.tsv', decimal=',', sep='\t', encoding='UTF-8')
df_l
```

```python
df_l['impact_P_Bio_Health_2'] = 100*df_l['impact_P_Bio_Health']/df_l['impact_P']
df_l['impact_P_Phy_Eng_2'] = 100*df_l['impact_P_Phy_Eng']/df_l['impact_P']
df_l
```

```python
df_l['topic_specialisation'] = 0
df_l['topic_specialisation_bio'] = 0
df_l['topic_specialisation_eng'] = 0
df_l.loc[df_l['impact_P_Bio_Health_2']>=50, 'topic_specialisation_bio'] = 1
df_l.loc[df_l['impact_P_Phy_Eng_2']>=50, 'topic_specialisation_eng'] = 1
df_l.loc[(df_l['impact_P_Bio_Health_2']>=50) | (df_l['impact_P_Phy_Eng_2']>=50), 'topic_specialisation'] = 1
df_l.drop(columns=['impact_P_Bio_Health_2', 'impact_P_Phy_Eng_2'], inplace=True)
df_l
```

```python
df_l['gini_production'] = None

for i in range(df_l.shape[0]):
    df_l.loc[i, 'gini_production'] = round(gini(df_l.loc[i ,['impact_P_Bio_Health', 'impact_P_Life_Earth', 'impact_P_Math_Comp', 'impact_P_Phy_Eng', 'impact_P_Soc_Hum']].tolist()), 3)

df_l
```
